### 4. 각각의 R, G, B에 대한 가우시안 분포를 정규화 히스토그램에 겹쳐서 그려봐라! 그리고 3차원(R,G,B축) 산포도를 그려라!

```python
def gaussian(x,mean,sigma):
    return (1/np.sqrt(2*np.pi*sigma**2)) * np.exp(-(x-mean)**2 / (2*sigma**2))
    
```

```python
# 새벽
from scipy.special import erf
plt.style.use('default')
plt.rcParams['figure.figsize'] = (6, 3)
plt.rcParams['font.size'] = 12
plt.rcParams['lines.linewidth'] = 5
x = np.arange(-12800,12800,100)
print(x.shape)
R_0_gaus =gaussian(x,r_0_avg,r_0_var)
G_0_gaus =gaussian(x,g_0_avg,g_0_var) 
B_0_gaus =gaussian(x,b_0_avg,b_0_var)  
hist0_R = cv2.calcHist([img0_norm],[0],None,[256],[0,256])
hist0_G = cv2.calcHist([img0_norm],[1],None,[256],[0,256])
hist0_B = cv2.calcHist([img0_norm],[2],None,[256],[0,256])

plt.plot(x,R_0_gaus,alpha=0.7,label='R_0')
plt.plot(x,G_0_gaus,alpha=0.7,label='G_0')
plt.plot(x,B_0_gaus,alpha=0.7,label='B_0')
plt.plot(x,hist0_R,alpha=0.8,label='hist0_R')
plt.plot(x,hist0_G,alpha=0.8,label='hist0_G')
plt.plot(x,hist0_B,alpha=0.8,label='hist0_B')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper left')
plt.show()
```

```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
def proecessed_data(data, number):
    file_list = sorted(os.listdir(f'data/{data}/'))
    if file_list[0] != '.DS_Store':
        number -= 1
    path = f'data/{data}/{file_list[number]}'
    img = cv2.imread(f'{path}',cv2.IMREAD_COLOR)
    img = cv2.resize(img, (256, 256))
    cv2.imwrite(f'porcessed_data/{data}/{file_list[number]}', img)

def load_data(data, number):
    file_list = sorted(os.listdir(f'porcessed_data/{data}/'))
    if file_list[0] != '.DS_Store':
        number -= 1
    path = f'porcessed_data/{data}/{file_list[number]}'
    img = cv2.imread(f'{path}',cv2.IMREAD_COLOR)
    return img
```

```python
data = []
y = []
for i in range(20):
    img=cv2.imread('./새벽/{}.png'.format(i))
    img = cv2.resize(img,(256,256))
    img = img.reshape(-1)
    data.append(img)
    y.append(0)
for i in range(20):
    img=cv2.imread('./아침/{}.png'.format(i))
    img = cv2.resize(img,(256,256))
    img = img.reshape(-1)
    data.append(img)
    y.append(1)
for i in range(20):
    img=cv2.imread('./오후/{}.png'.format(i))
    img = cv2.resize(img,(256,256))
    img = img.reshape(-1)
    data.append(img)
    y.append(2)
for i in range(20):
    img=cv2.imread('./밤/{}.png'.format(i))
    img = cv2.resize(img,(256,256))
    img = img.reshape(-1)
    data.append(img)
    y.append(3)
```

```python
import torch
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(data,y,test_size=0.2,random_state=7)
print(len(X_train),len(Y_train))
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
model1 = DecisionTreeClassifier(random_state=32)
model1.fit(X_train,Y_train)

model2 = SVC(kernel='poly',C=3,degree=3)
model2.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
pred = model1.predict(X_test)
print('The accuracy of the Decision Tree is: {:.3f}'.format(accuracy_score(pred,Y_test)))
pred1 = model2.predict(X_test)
print('The accuracy of the Support Vector Machine is: {:.3f}'.format(accuracy_score(pred1,Y_test)))
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import plot_confusion_matrix

labels = ['새벽', '아침', '오후']
disp = plot_confusion_matrix(model1, 
                             X_test, Y_test, 
                             display_labels=labels, 
                             cmap=plt.cm.Reds, 
                             normalize=None)
disp.ax_.set_title('DT Confusion Matrix');
```
