#### 1,2. sin1, sin2, sin3의 파형과 spectrum

```python
import numpy as np
import matplotlib.pyplot as plt

def sin_wave(amp, freq, time):
    return amp * np.sin(2*np.pi*freq*time)
```

```python
time = np.arange(0, 10, 0.001)
sin1 = sin_wave(4, 10, time)
sin2 = sin_wave(1, 5, time)
sin3 = sin_wave(10, 1, time)
```

```python
plt.figure(figsize=(12,5))
plt.plot(time, sin1, label=r"$\sin {20\pi} t$", color='red')
plt.plot(time, sin2, label=r"$2\sin {10\pi} t$", color='blue')
plt.plot(time, sin3, label=r"$4\sin {2\pi} t$", color='green')

plt.legend()
plt.grid()
plt.show()
```

```python
# 스펙트럼 of sin1
samplingFreq=100
ft1 = np.fft.fft(sin1)/len(sin1)
# ft1 = ft1[range(int(len(sin1)/2))]

t = np.arange(0,10,1/samplingFreq)
cnt1 = len(sin1)
values1 = np.arange(int(cnt1))
timeperiod1 = cnt1/samplingFreq
freq1 = values1/timeperiod1

plt.subplot(1,2,1)
plt.plot(freq1,ft1)

plt.subplot(1,2,2)
plt.specgram(ft1,Fs=100,cmap='rainbow')
```

```python
# 스펙트럼 of sin2
samplingFreq=1000
ft2 = np.fft.fft(sin2)/len(sin2)
# ft2 = ft2[range(int(len(sin2)/2))]
t = np.arange(0,10,1/samplingFreq)
cnt2 = len(sin2)
values2 = np.arange(int(cnt2))
timeperiod2 = cnt2/samplingFreq
freq2 = values2/timeperiod2

plt.subplot(1,2,1)
plt.plot(freq1,ft2)

plt.subplot(1,2,2)
plt.specgram(ft2,Fs=100,cmap='rainbow')
```

```python
# 스펙트럼 of sin3
samplingFreq=100
ft3 = np.fft.fft(sin3)/len(sin3)
# ft3 = ft3[range(int(len(sin3)/2))]
t = np.arange(0,10,1/samplingFreq)
cnt3 = len(sin3)
values3 = np.arange(int(cnt3))
timeperiod3 = cnt3/samplingFreq
freq3 = values3/timeperiod3

plt.subplot(1,2,1)
plt.plot(freq1,ft3)

plt.subplot(1,2,2)
plt.specgram(ft3,Fs=100,cmap='rainbow')
```

#### 3,4. sin1, sin2, sin3의 합성 파형 및 스펙트럼 그리기

```python
sin_sum = sin1 + sin2 + sin3
plt.figure(figsize=(12,5))
plt.plot(time, sin_sum)|
plt.grid()
plt.show()
```

```python
#스펙트럼 of sin_sum
samplingFreq=100
ft_sum = np.fft.fft(sin_sum)/len(sin_sum)
# ft_sum = ft_sum[range(int(len(sin_sum)/2))]
t = np.arange(0,10,1/samplingFreq)
cnt_sum = len(sin_sum)
values_sum = np.arange(int(cnt_sum))
timeperiod_sum = cnt_sum/samplingFreq
freq_sum = values_sum/timeperiod_sum

plt.subplot(1,2,1)
plt.plot(freq1,ft_sum)

plt.subplot(1,2,2)
plt.specgram(ft_sum,Fs=100,cmap='rainbow')
```

#### 5,6. sin1, sin2, sin3을 시간축에서 순차적으로 연결한 파형 및 스펙트럼 그리기

```python
sin_cat = np.concatenate((sin1,sin2,sin3))
print(sin_cat.shape)
time = np.arange(0, 30, 0.001)

plt.figure(figsize=(12,5))
plt.plot(time, sin_cat)
plt.grid()
plt.show()
```

```python
# 스펙트럼 of sin_cat
samplingFreq=100
ft_cat = np.fft.fft(sin_cat)/len(sin_cat)
# ft_cat = ft_cat[range(int(len(sin_cat)/2))]
t = np.arange(0,10,1/samplingFreq)
cnt_cat = len(sin_cat)
values_cat = np.arange(int(cnt_cat))
timeperiod_cat = cnt_cat/samplingFreq
freq_cat = values_cat/timeperiod_cat


plt.subplot(1,2,1)
plt.plot(freq_cat,ft_cat)

plt.subplot(1,2,2)
plt.specgram(ft_cat,Fs=100,cmap='rainbow')
```

```python
import numpy as np
import librosa, librosa.display 
import matplotlib.pyplot as plt
file = './hi.wav'

# load audio file with Librosa
sig, sr = librosa.load(file,sr=22050)
print(sig,sig.shape)
```

```python
plt.figure(figsize=(15,10))
librosa.display.waveshow(sig, sr, alpha=0.5)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

```python
fft = np.fft.fft(sig) /len(sig)

# 복소공간 값 절댓갑 취해서, magnitude 구하기
magnitude =fft 

# Frequency 값 만들기
f = np.linspace(0,sr,len(magnitude))

# 푸리에 변환을 통과한 specturm은 대칭구조로 나와서 high frequency 부분 절반을 날려고 앞쪽 절반만 사용한다.
# left_spectrum = magnitude[:int(len(magnitude))]
# left_f = f[:int(len(magnitude)/2)]

plt.figure(figsize=(15,10))
plt.plot(f,magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Power spectrum")
```
